```python
import numpy as np
import scipy.io as sio
import torch

from DAST_Network import DAST


def load_array(path, key):
    return sio.loadmat(path)[key]
train_dataset='train_dataset'

trainX = load_array(f"{train_dataset}/F001_window_size_trainX.mat", "train1X")
trainY = load_array(f"{train_dataset}/F001_window_size_trainY.mat", "train1Y").flatten()
testX = load_array(f"{train_dataset}/F001_window_size_testX.mat", "test1X")
testY = load_array(f"{train_dataset}/F001_window_size_testY.mat", "test1Y").flatten()
trainX.ndim, trainY.ndim, testX.ndim, testY.ndim
```

```python
import numpy as np
print("RUL 最大值:", np.max(trainY))
print("RUL 最小值:", np.min(trainY))
print("RUL 平均值:", np.mean(trainY))
print("RUL 中位数:", np.median(trainY))
print("RUL 标准差:", np.std(trainY))
```
